```python
import os
import pandas as pd
# 定义目录路径
processed_dir = 'F:\\quant_comp\\quant_comp\\future_L2\\processed_test'
drop_cols = ['WAP', 'LogReturn', 'OBI_L1', 'OBI_L2', 'Spread_L1', 'Spread_L2', 'DepthImbalance', 'Volatility', 'Momentum', 'TimeOfDay']
# 遍历所有日期文件夹
for date_folder in os.listdir(processed_dir):
    date_path = os.path.join(processed_dir, date_folder)
    if os.path.isdir(date_path):
        # 遍历每个日期文件夹中的所有parquet文件
        for file in os.listdir(date_path):
            if file.endswith('.parquet'):
                file_path = os.path.join(date_path, file)
                print(f"Processing: {file_path}")
                
                # 读取parquet文件
                df_temp = pd.read_parquet(file_path)
                
                # 删除指定列（如果存在）
                cols_to_drop = [col for col in drop_cols if col in df_temp.columns]
                if cols_to_drop:
                    df_temp = df_temp.drop(columns=cols_to_drop)
                    print(f"Dropped columns: {cols_to_drop}")
                
                # 覆盖原文件
                df_temp.to_parquet(file_path, index=False)
                print(f"Updated: {file_path}")

print("All files processed successfully!")
```
